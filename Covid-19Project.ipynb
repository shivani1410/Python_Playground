```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("C:/Users/shivani.mathur01/Desktop/Work/python/Chapter1/covid_19_india.csv")
df
```

```python
imputer=SimpleImputer(strategy='constant')
df2=pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
```

```python
df3=df2.groupby(['State/UnionTerritory','Date'])[['State/UnionTerritory','Date','Cured','Confirmed','Deaths']].sum().reset_index()
```

```python
df3.head(20)
```

```python
states=df3['State/UnionTerritory'].unique()
len(states)
```

```python
for i in range(0,len(states)):
    C=df3[df3['State/UnionTerritory']==states[i]].reset_index()
    plt.scatter(np.arange(0,len(C)),C['Confirmed'],color='b',label='Confirmed')
    plt.scatter(np.arange(0,len(C)),C['Cured'],color='g',label='Cured')
    plt.scatter(np.arange(0,len(C)),C['Deaths'],color='r',label='Deaths')
    plt.title(states[i])
    plt.xlabel('Days since first patient')
    plt.ylabel('No. of Cases')
    plt.legend()
    plt.show()
```

```python
df5=df2.groupby(['Date'])[[Confirmed','Cured','Deaths']].sum().reset_index()
df5
```

```python
K=df5
plt.scatter(np.arange(0,len(K)),K['Confirmed'],color='b',label='Confirmed')
plt.scatter(np.arange(0,len(K)),K['Cured'],color='g',label='Cured')
plt.scatter(np.arange(0,len(K)),K['Deaths'],color='r',label='Deaths')
plt.title('India')
plt.xlabel('Days since first patient')
plt.ylabel('No. of Cases')
plt.legend()
plt.show()
```
