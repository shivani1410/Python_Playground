```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)
#Reading Data
data=pd.read_csv("C:/Users/shivani.mathur01/Desktop/Work/python/files/headbrain.csv")
data.head()
```

```python
#Collecting X and Y
X=data["Head Size(cm^3)"].values
Y=data["Brain Weight(grams)"].values
```

```python
# Mean X and Y
mean_x=np.mean(X)
mean_y=np.mean(Y)

n=len(X)
nominator=0
demoninator=0
for i in range(n):
    nominator+=(X[i]-mean_x)*(Y[i]-mean_y)
    demoninator+=(X[i]-mean_x)**2
m=nominator/demoninator
c=mean_y-(m*mean_x)
print(m)
print(c)
```

```python
max_x=np.max(X)+100
min_x=np.min(X)-100
x=np.linspace(min_x,max_x,1000)
y=c+m*x
plt.plot(x,y, color='b',label='Regression Line')
plt.scatter(X,Y, color='r',label='Scatter')
plt.xlabel('Head Size')
plt.ylabel('Brain Size')
plt.legend()
plt.show()
```

```python
nom_R_sq=0
denom_R_sq=0
for i in range(n):
    y_pred=c+m*X[i]
    nom_R_sq+=(y_pred-mean_y)**2
    denom_R_sq+=(Y[i]-mean_y)**2
R_sq=nom_R_sq/denom_R_sq
print(R_sq)
```
